**making changes to the data**


```python
#df['total']=df['HP']+df['Attack']+df['Defense']+df['Sp. Atk']+df['Sp. Def']+df['Speed']
df['total']=df.iloc[:,4:10].sum(axis=1) #toplam bu sekildede yapılabilir
df.head(5)
```

```python
df=df.drop(columns=["total"])#sutunu siler
df.head(5)
```

```python
cols=list(df.columns)
df=df[cols[0:4]+[cols[-1]]+cols[4:12]]
df.head(5)
#sutunların sıralamasını degiştirebiliriz
```

**saving our data**

```python
df.to_csv('modified.csv',index="False")
#yaptıgımız degiklikleri csv olarak kaydedebiliriz
```

**Groupby**

```python
df=pd.read_csv('modified.csv')
```

```python
df.groupby(['Type 1']).mean(numeric_only=True).sort_values('Defense',ascending=False)
```

```python
df.groupby(['Type 1']).count()
```

```python
df.groupby('Generation').count()['Name']#jenerasyonlara göre poke sayıarı
```

```python
df.groupby('Legendary').mean(numeric_only=True)
```

```python
df.loc[df.groupby('Type 1')['total'].idxmax()]#type göre en iyi pokemonlar
```

```python
df.groupby('Type 1')['Attack'].mean().sort_values(ascending=False)#atak gucune göre
```

**working with large amounts of data**

```python
#buyuk veri setlerinde bellek kullanımı için chunksize kullanılır
#veriyi parçalara bölerek işlemi yapar
for df in pd.read_csv('Pokemon.csv', chunksize=100):
    fire_pokemon = df[df['Type 1'] == 'Fire']
    print(fire_pokemon)
```
