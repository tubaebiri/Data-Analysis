**YouTube verilerini analiz edilmektedir. Verilerin işlenmesi, görselleştirilmesi ve yorumlanması amacıyla çeşitli Python kütüphaneleri kullanılmıştır.**

```python
import pandas as pd
import numpy as np

df=pd.read_csv("/content/Global YouTube Statistics.csv",encoding="latin-1")
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.fillna({"Country": "Unknown"}, inplace=True)
```

```python
df.drop(columns=["Abbreviation", "Latitude", "Longitude"], inplace=True)
```

```python
turkey_channels = df[df["Country"] == "Turkey"]
print(turkey_channels[["Youtuber", "subscribers", "video views"]].head(10))
```

```python
entertainment_channels = df[df["category"] == "Entertainment"]
entertainment_channels.head(10)
```

```python
us_channels=df[df["Country"]=="United States"]
print(us_channels[["rank","Youtuber","channel_type"]].head(10))
```

```python
df.groupby("Country")["subscribers"].sum().sort_values(ascending=False).head(10)
```

```python
big_channels = df[df["subscribers"] > 10_000_000]
big_channels.head(5)
```

```python
popular_last_month = df[df["video_views_for_the_last_30_days"] > 100_000_000]
popular_last_month.head(5)
```

```python
old_channels = df[df["created_year"] < 2015]
old_channels.head(5)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
```

```python
# Ülkeye göre toplam abone sayısı
top_countries = df.groupby("Country")["subscribers"].sum().nlargest(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette="viridis")
plt.xlabel("Ülke")
plt.ylabel("Toplam Abone Sayısı")
plt.title("En Çok Aboneye Sahip 10 Ülke")
plt.xticks(rotation=45)
plt.show()
```

```python
# Kategoriye göre ortalama abone sayısı
avg_subs_by_category = df.groupby("category")["subscribers"].mean().nlargest(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=avg_subs_by_category.index, y=avg_subs_by_category.values, palette="coolwarm")
plt.xlabel("Kategori")
plt.ylabel("Ortalama Abone Sayısı")
plt.title("Kategoriye Göre Ortalama Abone Sayısı")
plt.xticks(rotation=45)
plt.show()
```

```python
top_30_days = df.nlargest(10, "video_views_for_the_last_30_days")
plt.figure(figsize=(12, 6))
sns.barplot(x=top_30_days["Youtuber"], y=top_30_days["video_views_for_the_last_30_days"], palette="plasma")
plt.title("Son 30 Gün İçinde En Çok İzlenen 10 Kanal")
plt.xticks(rotation=45)
plt.show()
```

```python
import plotly.express as px

# Ülkelere göre YouTuber sayısı
country_counts = df["Country"].value_counts().reset_index()
country_counts.columns = ["Country", "YouTuber_Count"]

fig = px.choropleth(
    country_counts,
    locations="Country",
    locationmode="country names",
    color="YouTuber_Count",
    hover_name="Country",
    title="Dünya Genelinde YouTuber Dağılımı",
    color_continuous_scale="Reds"
)

fig.show()
```
